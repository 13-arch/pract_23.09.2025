```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from random import randint, choice
import pandas as pd
import time
import logging


logging.basicConfig(
    filename="log.txt",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    encoding="utf-8"
)


def get_proxies():
    url = "https://free-proxy-list.net/"
    r = requests.get(url, timeout=10)
    soup = BeautifulSoup(r.text, "html.parser")
    proxies = []
    for row in soup.select("table#proxylisttable tbody tr"):
        cols = row.find_all("td")
        ip = cols[0].text.strip()
        port = cols[1].text.strip()
        anonymity = cols[4].text.strip().lower()
        https = cols[6].text.strip()
        if anonymity in ["anonymous", "elite proxy"] and https == "yes":
            proxies.append(f"{ip}:{port}")
    return proxies


def check_proxy(proxy):
    try:
        r = requests.get("https://httpbin.org/ip",
                         proxies={"http": f"http://{proxy}", "https": f"http://{proxy}"},
                         timeout=5)
        return r.status_code == 200
    except:
        return False


def create_driver(proxy):
    random_version = randint(120, 140)
    opts = Options()
    opts.add_argument(
        f"user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        f"AppleWebKit/537.36 (KHTML, like Gecko) Chrome/{random_version}.0.0.0 Safari/537.36"
    )
    opts.add_argument(f"--proxy-server=http://{proxy}")
    logging.info(f"Создан новый драйвер: Chrome/{random_version}, прокси {proxy}")
    return webdriver.Chrome(options=opts)

def keep_watts(text):
    return text.strip() if text and "Вт" in text else None

def keep_hertz(text):
    return text.strip() if text and "Гц" in text else None


all_proxies = get_proxies()
working_proxies = [p for p in all_proxies if check_proxy(p)]
logging.info(f"Найдено рабочих прокси: {len(working_proxies)}")

if not working_proxies:
    raise Exception("Нет рабочих прокси!")


df = pd.DataFrame(columns=["Name", "Type", "Min Power", "Max Power", "Min Freq", "Max Freq", "URL"])

iter = 2
request_count = 0
proxy = choice(working_proxies)
driver = create_driver(proxy)


for item in very_clear_links:
    if request_count > 0 and request_count % 15 == 0:
        driver.quit()
        proxy = choice(working_proxies)
        driver = create_driver(proxy)
        logging.info("Смена User-Agent и прокси каждые 15 запросов")

    try:
        driver.get(item)
        logging.info(f"Открыта ссылка: {item}")
        time.sleep(1)

        try:
            name = driver.find_element(By.XPATH, '/html/body/div[1]/main/div[2]/h1').get_attribute('textContent').strip()
        except:
            name = None

        try:
            type_ = driver.find_element(By.XPATH, '//*[@id="product-data"]/div[1]/section/div[1]/div[1]/div[1]/table/tbody/tr[1]/td').get_attribute('textContent').strip()
        except:
            type_ = None

        try:
            min_power_raw = driver.find_element(By.XPATH, '//*[@id="product-data"]/div[1]/section/div[1]/div[1]/div[2]/table/tbody/tr[1]/td').get_attribute('textContent')
        except:
            min_power_raw = None

        try:
            max_power_raw = driver.find_element(By.XPATH, '//*[@id="product-data"]/div[1]/section/div[1]/div[1]/div[2]/table/tbody/tr[2]/td').get_attribute('textContent')
        except:
            max_power_raw = None

        try:
            min_frequency_raw = driver.find_element(By.XPATH, '//*[@id="product-data"]/div[1]/section/div[1]/div[1]/div[3]/table/tbody/tr[1]/td').get_attribute('textContent')
        except:
            min_frequency_raw = None

        try:
            max_frequency_raw = driver.find_element(By.XPATH, '//*[@id="product-data"]/div[1]/section/div[1]/div[1]/div[3]/table/tbody/tr[2]/td').get_attribute('textContent')
        except:
            max_frequency_raw = None


        min_power = keep_watts(min_power_raw)
        max_power = keep_watts(max_power_raw)
        min_frequency = keep_hertz(min_frequency_raw)
        max_frequency = keep_hertz(max_frequency_raw)


        df.loc[iter] = [name, type_, min_power, max_power, min_frequency, max_frequency, item]
        iter += 1
        request_count += 1

    except Exception as e:
        logging.error(f"Ошибка при обработке {item}: {e}")
        driver.quit()
        proxy = choice(working_proxies)
        driver = create_driver(proxy)

driver.quit()
df.to_csv("car_audio.csv", index=False, encoding="utf-8-sig")
logging.info("Парсинг завершён, данные сохранены в car_audio.csv")
```

```python
import time, os
import pandas as pd
from random import randint
from concurrent.futures import ThreadPoolExecutor, as_completed
from queue import Queue

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC



def create_driver():
    random_version = randint(120, 140)
    ua = (
        f"Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        f"AppleWebKit/537.36 (KHTML, like Gecko) Chrome/{random_version}.0.0.0 Safari/537.36"
    )
    opts = Options()
    opts.add_argument(f"user-agent={ua}")
    opts.add_argument("--disable-gpu")
    opts.add_argument("--no-sandbox")
    opts.add_argument("--disable-dev-shm-usage")
    return webdriver.Chrome(options=opts)



def get_all_links_from_section(section_url):
    driver = create_driver()
    links = []
    try:
        driver.get(section_url)


        try:
            WebDriverWait(driver, 5).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="cookies-informer"]/div/div[3]/button[4]'))
            ).click()
        except:
            pass


        last_page = int(driver.find_element(
            By.XPATH,
            '/html/body/div[1]/main/div[3]/div[1]/div[53]/nav/a[last()]'
        ).get_attribute('textContent'))

        for page in range(1, last_page + 1):
            driver.get(f"{section_url}page{page}")
            time.sleep(2)
            items = driver.find_elements(By.CSS_SELECTOR, 'a[href*="/digital-caracoustics/"]')
            page_links = [el.get_attribute("href") for el in items]
            links.extend(page_links)
            print(f"Страница {page}: собрано {len(page_links)} ссылок")

    finally:
        driver.quit()
    return list(set(links))



def keep_watts(text):
    return text.strip() if text and "Вт" in text else None

def keep_hertz(text):
    return text.strip() if text and "Гц" in text else None



def parse_product(driver, url):
    driver.get(url)
    time.sleep(1)

    def safe_xpath(xpath):
        try:
            return driver.find_element(By.XPATH, xpath).get_attribute('textContent').strip()
        except:
            return None

    name = safe_xpath('/html/body/div[1]/main/div[2]/h1')
    type_ = safe_xpath('//*[@id="product-data"]//table/tbody/tr[1]/td')
    min_power_raw = safe_xpath('//*[@id="product-data"]//div[2]//table/tbody/tr[1]/td')
    max_power_raw = safe_xpath('//*[@id="product-data"]//div[2]//table/tbody/tr[2]/td')
    min_frequency_raw = safe_xpath('//*[@id="product-data"]//div[3]//table/tbody/tr[1]/td')
    max_frequency_raw = safe_xpath('//*[@id="product-data"]//div[3]//table/tbody/tr[2]/td')

    return {
        "Name": name,
        "Type": type_,
        "Min Power": keep_watts(min_power_raw),
        "Max Power": keep_watts(max_power_raw),
        "Min Freq": keep_hertz(min_frequency_raw),
        "Max Freq": keep_hertz(max_frequency_raw),
        "URL": url
    }



def worker(task_queue, worker_id):
    driver = create_driver()
    results = []
    try:
        while not task_queue.empty():
            try:
                idx, link = task_queue.get_nowait()
            except:
                break
            try:
                data = parse_product(driver, link)
                results.append(data)
                print(f"[Поток {worker_id}] {idx} обработан")
            except Exception as e:
                print(f"[Поток {worker_id}] Ошибка на {link}: {e}")
    finally:
        driver.quit()
    return results



section_url = "https://auto.1k.by/digital-caracoustics/"

print("Собираем все ссылки...")
all_links = get_all_links_from_section(section_url)
print(f"Найдено {len(all_links)} ссылок")


task_queue = Queue()
for i, link in enumerate(all_links, start=1):
    task_queue.put((i, link))

num_workers = 10
all_results = []

with ThreadPoolExecutor(max_workers=num_workers) as executor:
    futures = [executor.submit(worker, task_queue, wid) for wid in range(1, num_workers+1)]
    for f in as_completed(futures):
        all_results.extend(f.result())

df = pd.DataFrame(all_results)
df.to_csv("car_audio.csv", index=False, encoding="utf-8-sig")
print("Сохранено в car_audio.csv")

df.head()
```
